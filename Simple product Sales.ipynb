```python
import pandas as pd
```

```python
# Load CSV
df = pd.read_csv("simple_kaggle_dataset.csv")
```

```python
# read csv 
df
```

```python
# First 5 rows
print("First 5 Rows")
print(df.head())
```

```python
# Column names
print("\nColumns")
print(df.columns)
```

```python
# Revenue column
df['revenue'] = df['Quantity_Sold'] * df['Price']
```

```python
# Total revenue
total_revenue = df["revenue"].sum()
print("\nTotal Revenue:")
print(total_revenue)
```

```python
# Revenue by category
category_revenue = df.groupby('Category')['revenue'].sum()
highest_category = category_revenue.idxmax()
print("\nHighest Revenue Category:")
print(highest_category)
```

```python
# Average rating
avg_rating = (
    df.groupby("Category")["Rating"]
    .mean()
)

print("\nAverage Customer Rating:")
print(avg_rating)
```

```python
import matplotlib.pyplot as plt
# Revenue chart
plt.figure(figsize=(8,5))
category_revenue.plot(kind="bar")
plt.title("Revenue per Product Category")
plt.tight_layout()
plt.savefig("revenue_per_category.png")
plt.close()
```

```python
# Rating chart
plt.figure(figsize=(8,5))
avg_rating.plot(kind="bar", color="orange")
plt.title("Average Customer Rating")
plt.tight_layout()
plt.savefig("average_rating_per_category.png")
plt.close()
```
